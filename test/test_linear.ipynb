```python
import sys 
sys.path.append('..')
from projet_etu import *
import numpy as np 
import matplotlib as plt 
from mltools import *
from utils import *
```

```python
Xl, y = gen_arti(epsilon=0.1)
print(Xl.shape, y.shape)
plt.figure()
plot_data(Xl,y)


## Visualisation de la fonction de coût en 2D
plt.figure()
plt.show()
```

```python
linear = Linear(2, 1)
mse = MSELoss()
```

```python
# Boucle d'apprentissage
for i in range(1000):
    # Forward pass 
    y_pred = linear.forward(Xl)
    loss = mse.forward(y, y_pred)

    # Backward pass
    delta = mse.backward(y, y_pred)
    linear.backward_update_gradient(Xl, delta)

    # Mise à jour des paramètres
    linear.update_parameters()
    linear.zero_grad()

    # Affichage de la perte
    if i % 10 == 0:
        print(f"Étape {i}, perte : {loss.mean()}")
```

```python
plot_frontiere(Xl, lambda x: np.sign(linear.forward(x)), step=100)
plot_data(Xl, y)
```
